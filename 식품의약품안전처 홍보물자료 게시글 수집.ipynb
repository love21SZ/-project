```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs

all_data = []
for page_no in range(1,10):
    url= f'https://www.mfds.go.kr/brd/m_231/list.do?page={page_no}&srchFr=&srchTo=&srchWord=&srchTp=&itm_seq_1=0&itm_seq_2=0&multi_itm_seq=0&company_cd=&company_nm='
    response = requests.get(url)
    html = bs(response.text,"html.parser")
    
    title_list = []
    for i in range(1,11):
        post_no = i
        title = html.select(f"#content > div.bbs_list01 > ul > li:nth-child({post_no}) > div.center_column > a")
        title_N = title[0].text.strip()
        title_list.append(title_N)
    
    part_list = []   
    for i in range(1,11):
        post_no = i
        part = html.select(f"#content > div.bbs_list01 > ul > li:nth-child({post_no}) > div.center_column > div > p:nth-child(1)")
        part_p = part[0].text.strip().split(" | ")[-1]
        part_list.append(part_p)
    date_list = []    
    for i in range(1,11):
        post_no = i
        date = html.select(f"#content > div.bbs_list01 > ul > li:nth-child({post_no}) > div.right_column")
        date_a = date[0].text.strip().split(" | ")[-1]
        date_list.append(date_a)
        
    riew_list = []   
    for i in range(1,11):
        post_no = i
        riew =html.select (f"#content > div.bbs_list01 > ul > li:nth-child({post_no}) > div.center_column > div > p:nth-child(2)")
        riew_r = riew[0].text.strip().split(" | ")[-1]
        riew_list.append(riew_r)
        
    df = pd.DataFrame(zip(title_list,part_list,date_list,riew_list), columns = ["제목","담당부서","날짜","조회수"])
    all_data.append(df)
```

```python
df_dd = pd.concat(all_data)
df_dd
```
