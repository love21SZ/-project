```python
df = ai.melt(id_vars=ai.columns[:2],var_name ="연도",value_name="분율")
df.shape
```

```python
plt.figure(figsize=(5,5))
sns.barplot(data = df , x = "성별", y = "분율")
```

```python
table = df.pivot_table(index = "연도", columns = "연령대", values = "분율")
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(table.T, annot = True , fmt = ",.1f",cmap="Blues")
```

```python
plt.figure(figsize=(10,3))
sns.catplot(data=df, x ="연령대",y="분율")
```

```python
plt.figure(figsize=(15,3))
sns.boxplot(data = df , x = "연도", y = "분율")
```
